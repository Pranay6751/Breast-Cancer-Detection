```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

from keras.models import Sequential
from keras.layers import Dense
```

```python
scale=StandardScaler()
feature=scale.fit_transform(feature)
```

```python
#500 traning 

x_train=feature[:500]
y_train=label[:500]

#35 validation

x_val=feature[500:535]
y_val=label[500:535]

#34 testing

x_test=feature[535:]
y_test=label[535:]

```

```python

model = Sequential()  # empty stack for neurons         # we use dense layer because all the neurons in one dimension

model.add(Dense(15,activation='relu',input_dim=30))   #first hidden layer with  15 neurons and it contains 30 neurons in input layer

model.add(Dense(15,activation='relu'))                 # second hidden layer

model.add(Dense(1,activation='sigmoid'))             # output layer with one neuron 
 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])  # during propagating forward and backward adam algoritm optimizes  these weights to get more accuracy 
```

```python
model.fit(x_train,y_train,epochs=10,validation_data=(x_val,y_val))    # Training and validating the data
```

```python
model.evaluate(x_test,y_test)
```

```python
for i in range(34):
  sample = x_test[i]
  sample = np.reshape(sample, (1,30))

  if (model.predict(sample)[0][0] > 0.5):
    print("-Benign")
  else:
    print("-Malignant")

  if (y_test[i] == 1):
    print("*Banign")
  else:
    print("*Melignant")
  print("-----------")
```
